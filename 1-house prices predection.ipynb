## treat the outliers : 

#### hot rooms

```python
uv = np.percentile(df.n_hot_rooms,[99])[0]
```

```python
df[(df.n_hot_rooms>uv)]
```

```python
#d ai 
df.loc[df.n_hot_rooms > 3*uv, 'n_hot_rooms'] = uv *1.5
```

```python
plt.figure(figsize=(5,4))
plt.scatter(x='n_hot_rooms' , y='price' , data=df ,s=4)
```

#### rainfall

```python
lv = np.percentile(df.rainfall,[1])[0]
print(lv)
```

```python
df[(df.rainfall< lv)]
```

```python
df.loc[df.rainfall< 0.3*lv, 'rainfall'] = 0.3*lv
```

```python
plt.scatter(x='rainfall' , y='price' , data=df ,s=4)
```

## Machine Learning

### linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
y= df['price']
```

```python
X=df[['room_num']]
```

```python
model = LinearRegression()
```

```python
#this is a simple model , with one feture , one label . (x1,y)
model.fit(X,y)
```

```python
print(model.intercept_, model.coef_)
```

```python
# to understand  more about LR model : 
help(model)
```

```python
model.predict([[7]])
```

```python
model.predict(X)
```

```python
#seeing the score of the model
from sklearn.metrics import r2_score
r2_score( y, model.predict(X))
```

#### Do it with multuple varaibles

```python
X_multi = df.drop("price",axis=1)
```

```python
y_multi = df['price']
```

```python
lm3 = LinearRegression()
```

```python
lm3.fit(X_multi, y_multi)
```

```python
print(lm3.intercept_, lm3.coef_)
```

#### train test split 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X_multi , y_multi , test_size=0.2 , random_state=0)
```

```python
print(X_train.shape , X_test.shape , y_train.shape , y_test.shape) 
```

```python
lm_a= LinearRegression()
```

### fitting

```python
lm_a.fit(X_train,y_train)
```

```python
y_train_a = lm_a.predict(X_train)
```

```python
y_test_a = lm_a.predict(X_test)
```

## Result / R score output

```python
from sklearn.metrics import r2_score

```

```python

# #existing value 'y' , #    the predicted values 'y'of training sample

r2_score(y_train , y_train_a) 
```

```python
#.. of testing sample 
r2_score(y_test , y_test_a)
```
